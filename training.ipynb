```python
import numpy as np
import time

# 문제 설정
# 2,000만 개의 요소를 가진 1차원 데이터셋을 생성합니다.
# 엑셀 시트로 치면 2,000만 개의 행을 가진 단일 열 데이터라고 생각할 수 있습니다.
data_size = 20_000_000
scalar_a = 0.5
scalar_b = 10

# --------------------
# 1. 일반 Python 리스트와 반복문 사용
# --------------------
print("1. 일반 Python 리스트로 연산 시작...")
python_data = list(range(data_size))
start_time_python = time.time()

result_python = []
for x in python_data:
    result_python.append(scalar_a * x + scalar_b)

end_time_python = time.time()
print(f"일반 Python 연산 소요 시간: {end_time_python - start_time_python:.4f} 초")
print("-" * 30)

# --------------------
# 2. numpy 배열과 벡터화된 연산 사용
# --------------------
print("2. numpy 배열로 연산 시작...")
numpy_data = np.arange(data_size)
start_time_numpy = time.time()

# numpy의 벡터화된 연산은 반복문 없이 한 줄로 처리됩니다.
result_numpy = scalar_a * numpy_data + scalar_b

end_time_numpy = time.time()
print(f"numpy 연산 소요 시간: {end_time_numpy - start_time_numpy:.4f} 초")
print("-" * 30)

# --------------------
# 결과 비교
# --------------------
# 두 결과가 동일한지 확인합니다.
if np.allclose(result_python, result_numpy):
    print("두 결과가 동일합니다.")
else:
    print("결과가 다릅니다.")

# 요약
print(f"numpy가 {(end_time_python - start_time_python) / (end_time_numpy - start_time_numpy):.2f}배 빠릅니다.")
```

```python
sales = np.array([100, 150, 200, 120, 180])
costs = np.array([80, 120, 150, 90, 140])

profits = sales - costs
print(profits)

mean_profit = np.mean(profits)
median_profit = np.median(profits)
vairance_profit = np.var(profits)
print(f"평균 이익: {mean_profit}, 중앙값 이익: {median_profit}, 분산: {vairance_profit}")
```

```python
import numpy as np

# A, B, C 캠페인의 노출 수, 클릭 수, 전환 수
campaign_data = np.array([
    [1000000, 5000, 100],  # A 캠페인: Impressions, Clicks, Conversions
    [1200000, 4800, 150],  # B 캠페인
    [800000, 6400, 120]   # C 캠페인
])

CTR_np = campaign_data[:, 1] / campaign_data[:, 0] * 100  # 클릭률(CTR) 계산
CVR_np = campaign_data[:, 2] / campaign_data[:, 1] * 100  # 전환율(CVR) 계산

print(CTR_np)
print(CVR_np)
print(campaign_data[np.where(CVR_np == max(CVR_np))[0][0]])
```
